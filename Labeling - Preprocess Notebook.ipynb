# Pre-process

## Tokenzize sentences

```python
sentences_data = []
for index, row in raw_data.iterrows():
    review = row['review']
    separated_sentences = nltk.sent_tokenize(review)
    sentences_data.extend(separated_sentences)
data = pd.DataFrame({"review": sentences_data})
```

```python
data
```

## Remove special symbols

```python
def clean_text(text):
    text = re.sub('[^A-Za-z]+', ' ', text)
    return text
```

```python
data['cleaned_review'] = data['review'].apply(clean_text)
data.head()
```

## POS tagging

```python
pos_dict = {'J':wordnet.ADJ, 'V':wordnet.VERB, 'N':wordnet.NOUN, 'R':wordnet.ADV}
def token_stop_pos(text):
  tags = pos_tag(word_tokenize(text))
  newlist = []
  for word, tag in tags:
      if word.lower() not in set(stopwords.words('english')):
        newlist.append(tuple([word, pos_dict.get(tag[0])]))
  return newlist
```

```python
data['pos_tagged'] = data['cleaned_review'].apply(token_stop_pos)
data.head()
```

## Lemmatize

```python
wordnet_lemmatizer = WordNetLemmatizer()
def lemmatize(pos_data):
  lemma_rew = " "
  for word, pos in pos_data:
    if not pos:
        lemma = word
        lemma_rew = lemma_rew + " " + lemma
    else:
        lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
        lemma_rew = lemma_rew + " " + lemma
  return lemma_rew
```

```python
data['lemma'] = data['pos_tagged'].apply(lemmatize)
data.head()
```

## Tokenize lemmatized sentences

```python
def lemma_word_tokenize(text):
  return word_tokenize(text)
```

```python
data['lemma_words'] = data['lemma'].apply(lemma_word_tokenize)
data.head()
```

## Sentiment rule based labeling

```python
def sentiwordnetanalysis(pos_data):
  sentiment = 0
  tokens_count = 0
  for word, pos in pos_data:
      if not pos:
          continue
      lemma = wordnet_lemmatizer.lemmatize(word, pos=pos)
      if not lemma:
          continue
      synsets = wordnet.synsets(lemma, pos=pos)
      if not synsets:
          continue
      # Take the first sense, the most common
      synset = synsets[0]
      swn_synset = swn.senti_synset(synset.name())
      sentiment += swn_synset.pos_score() - swn_synset.neg_score()
      tokens_count += 1
      # print(swn_synset.pos_score(),swn_synset.neg_score(),swn_synset.obj_score())
  if not tokens_count:
      return None
  if sentiment>0:
      return "positive"
  if sentiment==0:
      return "neutral"
  else:
      return "negative"
```

```python
data['polarity'] = data['pos_tagged'].apply(sentiwordnetanalysis)
data.head()
```
